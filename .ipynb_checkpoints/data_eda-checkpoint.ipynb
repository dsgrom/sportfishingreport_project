```python
#import packages and set wd to sportfishingreport_project folder
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

os.chdir("C:/Users/Kevin/Documents/Projects/Portfolio/sportfishingreport_project")

#import 2019 data
df19_raw = pd.read_csv("sfr_data_cleaned_2019.csv")
df19 = df19_raw.copy()

#import 2020 and 2021 data, split into two separate dataframes
df20_raw = pd.read_csv("sfr_data_cleaned_v2.csv")
df20 = df20_raw.copy()
df21 = df20.loc[df20['year'] == 2021, :]
df20 = df20.loc[df20['year'] == 2020, :]
```

```python
#Dataset sample
df20.head()
```

```python
#Dimensions of each year dataset and column names
print('2019 DF dimensions: \n', df19.shape, '\n', sep = '')

print('\n2020 DF dimensions: \n', df20.shape, '\n', sep = '')

print('\n2021 DF dimensions: \n', df21.shape, '\n', sep = '')

print('\nColumn names: \n', df19.columns, sep = '')
```

```python
#2019 vs 2020 top 5 species by catch, 2020 catch should be lower due to covid
print(df19.groupby(['year', 'catch_species_general']).agg(catch_count = ('catch_count', 'sum')) \
    .sort_values(['catch_count'], ascending = False)[0:10])

print(df20.groupby(['year', 'catch_species_general']).agg(catch_count = ('catch_count', 'sum')) \
    .sort_values(['catch_count'], ascending = False)[0:10])
```

```python
#lists of top species for each year
top_species_19 = df19.groupby(['catch_species_general']).agg(catch_count = ('catch_count', 'sum')) \
    .sort_values(['catch_count'], ascending = False)[0:10].index
top_species_20 = df20.groupby(['catch_species_general']).agg(catch_count = ('catch_count', 'sum')) \
    .sort_values(['catch_count'], ascending = False)[0:10].index

#ranked dictionary for sorting to keep order of top 10 species
custom_sort_19 = {'rockfish':0, 'whitefish':1, 'bass':2, 'tuna':3, 'crab':4, \
                   'sculpin':5, 'sanddab':6, 'yellowtail':7, 'lingcod':8, 'salmon':9}
custom_sort_20 = {'rockfish':0, 'whitefish':1, 'bass':2, 'tuna':3, 'sanddab':4, \
                   'sculpin':5, 'bonito':6, 'crab':7, 'yellowtail':8, 'dorado':9}


top19 = df19[df19.catch_species_general.isin(top_species_19)].pivot_table( \
               values = 'catch_count', \
               index = 'catch_species_general', \
               columns = 'month', \
               aggfunc = np.sum, \
               fill_value = 0).sort_values(by = 'catch_species_general', key = lambda x: x.map(custom_sort_19))

top20 = df20[df20.catch_species_general.isin(top_species_20)].pivot_table( \
               values = 'catch_count', \
               index = 'catch_species_general', \
               columns = 'month', \
               aggfunc = np.sum, \
               fill_value = 0).sort_values(by = 'catch_species_general', key = lambda x: x.map(custom_sort_20))


#catch count for top 10 species of each year, by month
print('Catch count for top 10 species of 2019 by month:\n', top19, sep = '')

print('\n\n\nCatch count for top 10 species of 2020 by month:\n', top20, sep = '')
```

```python
#monthly 
#this is the graph that I want, but it's way too dense to be valuable
#df19.loc[df19['catch_species_general'] == 'rockfish'].plot(kind = 'line', x = 'date2', y = 'catch_count', figsize = (100, 30))
import matplotlib.axes as ax

top19_transpose = top19.transpose()
top20_transpose = top20.transpose()

#2019 plot
plt.figure(1, figsize = (30, 20))
plt.subplot(2,1,1)
plt.plot(top19_transpose.index, 'rockfish', data = top19_transpose, linewidth = 6, label = 'Rockfish')
plt.plot(top19_transpose.index, 'whitefish', data = top19_transpose, linewidth = 6, label = 'Whitefish')
plt.plot(top19_transpose.index, 'bass', data = top19_transpose, linewidth = 6, label = 'Bass')
plt.plot(top19_transpose.index, 'tuna', data = top19_transpose, linewidth = 6, label = 'Tuna')
plt.plot(top19_transpose.index, 'yellowtail', data = top19_transpose, linewidth = 6, label = 'Yellowtail')
plt.legend(fontsize = '18')
plt.xlabel('Month', fontsize = '18')
plt.ylabel('Fish Count', fontsize = '18')
plt.title('2019 Fish Counts by Month', fontsize = '24')
#somehow need to figure out how to manipulate axes

plt.subplot(2,1,2)
plt.plot(top20_transpose.index, 'rockfish', data = top20_transpose, linewidth = 6, label = 'Rockfish')
plt.plot(top20_transpose.index, 'whitefish', data = top20_transpose, linewidth = 6, label = 'Whitefish')
plt.plot(top20_transpose.index, 'bass', data = top20_transpose, linewidth = 6, label = 'Bass')
plt.plot(top20_transpose.index, 'tuna', data = top20_transpose, linewidth = 6, label = 'Tuna')
plt.plot(top20_transpose.index, 'yellowtail', data = top20_transpose, linewidth = 6, label = 'Yellowtail')
plt.legend(fontsize = '18')
plt.xlabel('Month', fontsize = '18')
plt.ylabel('Fish Count', fontsize = '18')
plt.title('2020 Fish Counts by Month', fontsize = '24')
```

```python
#Most boats
#2020 and 2019 into into 1 dataframe
df = df20.append(df19)

#list for iterating and initialize result df
socal_locations = ['san diego', 'orange', 'los angeles', 'ventura coast']
boat_counts = pd.DataFrame(columns = ['location', '2019', '2020'])

for region in socal_locations:
    #record number of unique boat names for each location for each year
    num_boats19 = df.loc[(df['location'] == region) & (df['year'] == 2019)].boat_name.unique().shape[0]
    num_boats20 = df.loc[(df['location'] == region) & (df['year'] == 2020)].boat_name.unique().shape[0]
    
    #temp df with values for this row
    temp = pd.DataFrame([[region, num_boats19, num_boats20]], \
                        columns = ['location', '2019', '2020'])
    
    boat_counts = boat_counts.append(temp)
    
boat_counts.set_index('location', inplace = True)
print('Number of boats by location:\n', boat_counts, sep = '')
```

```python
#pd.set_option('display.max_rows', None)

#which region has the most anglers? anglers per boat?

#group by date first to get rid of duplicate rows with individual species, since not interested in catch right now
#in this df, each line is number of anglers on each trip of the day
df_anglers = df.groupby(['year', 'date2', 'location', 'boat_name', 'trip_num'], as_index = False) \
               .agg(anglers = ('anglers', 'mean'))

df_anglers = df_anglers[df_anglers['location'].isin(['san diego', 'orange', 'los angeles', 'ventura coast'])]

#group new df by total over the year, rounding to eliminate some minor errors in the data
#in this df, each line is number of anglers each boat has taken in the year
df_anglers2 = df_anglers.groupby(['year', 'location', 'boat_name'], as_index = False) \
                        .agg(anglers = ('anglers', 'sum')).round()

#in this df, each line is aggregate number of anglers in each region each year
df_anglers_location = df_anglers2.groupby(['year', 'location'], as_index = False) \
                                 .agg(anglers = ('anglers', 'sum')).round()

#pivoted for clarity
print(df_anglers_location.pivot_table(values = 'anglers', index = 'location', columns = 'year', aggfunc = np.sum))


#pd.reset_option("display.max_rows")
```

```python
#per angler?
#need to aggregate catch and anglers by trip?

#aggregate catch and avg anglers on catch_species_general first
#each line is all of one generalized species caught by each boat per trip per day
#i.e. new del mar has 2 trips per day, each trip might have caught 1 bass, 2 lines each saying 1 bass
df_species_gen = df.groupby(['year', \
                             'date2', \
                             'location', \
                             'boat_name', \
                             'trip_num', \
                             'catch_species_general'], as_index = False) \
                   .agg(anglers = ('anglers', 'mean'), catch_count = ('catch_count', 'sum'))

#aggregate catch and anglers by location, for catch per angler calculation, rounded for simplicity
df_cpa = df_species_gen.groupby(['year', 'location', 'catch_species_general'], as_index = False) \
                                     .agg(anglers = ('anglers', 'sum'), catch_count = ('catch_count', 'sum')).round()

#2019 data
df_cpa19 = df_cpa.copy().loc[(df_cpa['location'].isin(['san diego', 'los angeles', 'orange', 'ventura coast'])) & \
                 (df_cpa['year'] == 2019) & \
                 (df_cpa['catch_species_general'].isin(top_species_19))]

#2020 data
df_cpa20 = df_cpa.copy().loc[(df_cpa['location'].isin(['san diego', 'los angeles', 'orange', 'ventura coast'])) & \
                 (df_cpa['year'] == 2020) & \
                 (df_cpa['catch_species_general'].isin(top_species_20))]
#print(df_cpa)

df_cpa19['cpa'] = df_cpa19['catch_count']/df_cpa19['anglers']
df_cpa20['cpa'] = df_cpa20['catch_count']/df_cpa20['anglers']

print('2019 catch per angler:\n', df_cpa19.drop(['anglers', 'catch_count'], axis = 1), sep = '')
print('\n\n\n2020 catch per angler:\n', df_cpa20.drop(['anglers', 'catch_count'], axis = 1), sep = '')


#NOTE: >2500 SANDDAB IS NOT THAT UNCOMMON FOR LA/VENTURA, ACCORDING TO TRAVIS THEY PILE UP ON THOSE BANKS AND PPL JUST
#MOP THEM UP. ANGLERS WILL BRING OUT BUCKETS TO CARRY ALL OF THEM (wild)
```
